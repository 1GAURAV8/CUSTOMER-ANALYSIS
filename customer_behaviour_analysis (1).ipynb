### Q7. What are the top 3 Purchased products within each category?

### Q1. Which customer segments contribute disproportionately high revenue compared to their population size?

### Segment - no of previous purchase

```python
query = '''   SELECT customer_segment, COUNT(*) AS count_of_each_segment, SUM(revenue) as segment_wise_revenue
		FROM (SELECT customer_id, previous_purchases, SUM(purchase_amount) as revenue, 
																		CASE 
																				WHEN previous_purchases < 11 THEN 'NEW'
																				WHEN previous_purchases BETWEEN 11 AND 25 THEN 'RETURNING'
																				ELSE 'LOYAL'
																		END AS customer_segment
				  FROM shopping_cleaned
                  GROUP BY customer_id, previous_purchases) AS segment
		GROUP BY customer_segment 
        ORDER BY segment_wise_revenue DESC'''

segment_wise_revenue = pd.read_sql(query, engine)
segment_wise_revenue
```

### Q2 Spending Segments of Customers
### SPENDS - SUM(PURCHASE_AMOUNT)

### SEGMENT OF CUSTOMERS: 1. HIGH - THOSE WHO SPENDS BETWEEN 75 AND 100
###                       2. ABOVE AVERAGE - THOSE WHO SPENDS BETWEEN 50 AND 75
###                       3. AVERAGE - THOSE WHO SPENDS BETWEEN 25 AND 50
###                       4. LOW - THOSE WHO SPENDS LESS THAN 25

### Segment - their total purchase amounts

```python
query = ''' WITH  spenders AS (SELECT customer_id, frequency_purchase_in_days, SUM(purchase_amount) as spending, COUNT(*) as frequency
        FROM shopping_cleaned
        GROUP BY customer_id, frequency_purchase_in_days) 
        
        SELECT 
													CASE 
															WHEN spending < 25 THEN 'LOW' 
															WHEN spending BETWEEN 25 AND 50 THEN 'AVERAGE'
															WHEN spending BETWEEN 50 AND 75 THEN 'ABOVE AVERAGE'
															WHEN spending BETWEEN 75 AND 100 THEN 'HIGH'
															ELSE 'EXTRA HIGH'
													END AS spending_segment,
				 SUM(spending) AS segment_revenue, COUNT(*) AS customer_count
        FROM spenders
        GROUP BY spending_segment
        ORDER BY segment_revenue DESC
          ''' 

Spending_segment = pd.read_sql(query, engine)
Spending_segment
```

```python
plt.figure(figsize = (9,6))
bars = plt.bar(Spending_segment['spending_segment'], Spending_segment['segment_revenue'], color = ['#90DBF4', '#FFD6A5', '#CDB4DB'] )
plt.xlabel('spending_segment')
plt.ylabel('segment_revenue')
plt.title('Revenue generated per Spending_segment')


legend_labels = [
    f"{seg} (Count_of_Customers: {cnt})"
    for seg, cnt in zip(Spending_segment['spending_segment'], Spending_segment['customer_count'])
]
plt.legend(bars, legend_labels, title="Segments", loc='best')  #title- legend_title, bars- output of plt.bar, legen_labes- list contiaining labels.

plt.bar_label(bars, fmt='%.0f', padding=3)
plt.tight_layout()
plt.show()
```

```python
query = ''' WITH  spenders AS (SELECT customer_id, age_group, frequency_purchase_in_days, SUM(purchase_amount) as spending, COUNT(*) as frequency
        FROM shopping_cleaned
        GROUP BY customer_id, frequency_purchase_in_days, age_group) 
        
        SELECT 
													CASE 
															WHEN spending < 25 THEN 'LOW' 
															WHEN spending BETWEEN 25 AND 50 THEN 'AVERAGE'
															WHEN spending BETWEEN 50 AND 75 THEN 'ABOVE AVERAGE'
															WHEN spending BETWEEN 75 AND 100 THEN 'HIGH'
															ELSE 'EXTRA HIGH'
													END AS spending_segment,
				 SUM(spending) AS segment_revenue, COUNT(*) AS customer_count, age_group
        FROM spenders
        GROUP BY  age_group, spending_segment
        ORDER BY segment_revenue DESC
          ''' 
Age_wise_Spending_segment = pd.read_sql(query, engine)
Age_wise_Spending_segment
```

### Q4.  Calculate median purchase amount per gender.

```python
query = ''' WITH A as (SELECT gender, purchase_amount, 
        ROW_NUMBER() OVER(PARTITION BY gender ORDER BY purchase_amount) as Rnk,
        COUNT(*) OVER(PARTITION BY gender) AS N
        FROM shopping_cleaned
        GROUP BY  gender, purchase_amount
        ORDER BY purchase_amount)
        
        SELECT gender, 
						CASE 	
							WHEN N%2 != 0 THEN (SELECT AVG(purchase_amount) FROM A WHERE Rnk = ((N+1)/2))
							ELSE (SELECT AVG(purchase_amount) FROM A WHERE Rnk IN ( (N/2), ((N+1)/2) + 1 )  )
							END as median_purchase_amount
		FROM A
        GROUP BY gender
        LIMIT 2 '''

median_amount = pd.read_sql(query, engine)
median_amount
```

```python
plt.figure(figsize = (9,6))
bars = plt.bar(median_amount['gender'], median_amount['median_purchase_amount'], color = ['#90DBF4', '#FFD6A5', '#CDB4DB'] )
plt.xlabel('gender')
plt.ylabel('median_purchase_amount')
plt.title('Median Purchase Amount per Gender')


legend_labels = [
    f"{seg} (Count_of_Customers: {cnt})"
    for seg, cnt in zip(median_amount['gender'], median_amount['median_purchase_amount'])
]
plt.legend(bars, legend_labels, title="Segments", loc='best')  #title- legend_title, bars- output of plt.bar, legen_labes- list contiaining labels.

plt.bar_label(bars, fmt='%.0f', padding=3)
plt.tight_layout()
plt.show()
```
